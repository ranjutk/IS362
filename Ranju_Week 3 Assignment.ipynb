## Bike Ride From Brooklyn Bridge, Manhattan to Key West, Florida

This is a log of my cycling from Lower point of Manhattan to the southernmost point of United States.
I will be logging my daily distance covered.
Total distance as per Google Maps is 1476 miles.

```python
# Import required libraries
from pandas import DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mp
import random
import numpy as np
import pprint

# Enable inline plotting
%matplotlib inline
```

My first two entries at the end of day are 55 and 120, indicating that I rode your bike 55 miles on day 1 and 65 miles on day 2.
Logging them and assigning known parameters.

```python
total_dist=1476
#Logging daily miles traveled
d1=55
d2=65
#Let's 
#Lets find average distance covered from the frist two days so we can get apporximate days required needed to reach Key West
avg_dist=(d1+d2)/2
# we are estimating that I travel 60 miles per day. This means I will take total_days to complete the trip. We are rounding the distance to two decimal points.
total_days= int(round(total_dist/avg_dist))
# As two days are over we will find the remaining days.

print ("Total Distance is from Manhattan to Key West is: ", total_dist, "miles")
print ("Average distance intend to travel per day is: ", avg_dist ,"miles")
print ("So, the total days of the journey will be: ",total_days,"days")
```

### Let create the log for distance covered each day. Best approach will be to create a list and link days to the miles covered.

```python
distlist= [d1,d2]
daylist= ["day 1","day 2"]

#Miles covered can be generated using random library and random range function between the first two day values
i=0
day=2
while (day-2) != (total_days-2): # as we already have 2 day data
    miles_random=random.randrange(d1,d2)
    day= day+1
    distlist.append(miles_random)
    daylist.append("day "+str(day))
    milesum=sum(distlist)
    if milesum >= total_dist:
        break
print("Distance covered each day")
dataset= list(zip(daylist,distlist))
#print (dataset)
count=0
while count < day:    
    print(daylist[count],":",distlist[count],"miles")
    count=count+1
```

### Now in panda series it will be

```python
df=pd.DataFrame(data=dataset, columns=['Day', 'Miles Covered'])
df
```

### Let us plot this in chart

```python
df['Miles Covered'].plot()

maxvalue_dist = df['Miles Covered'].max()

maxday = df['Day'][df['Miles Covered'] == df['Miles Covered'].max()].values

text = str(maxvalue_dist) + " - " + maxday
plt.annotate(text, xy=(1, maxvalue_dist), xytext=(8, 0), 
                 xycoords=('axes fraction', 'data'), textcoords='offset points')
#print("The most popular name")
df[df['Miles Covered'] == df['Miles Covered'].max()]
```

### Let find cumulative distance covered after each day.

```python
cumilative=[]
cum=0
ct=0
while cum <= total_dist:
    cum=cum+distlist[ct]
    #print (daylist[ct],":",cum,"miles")
    cumilative.append(cum)
    ct=ct+1
    if ct >= int(total_days):
        break
```

### Making it in panda series

```python
dataset_cumilative=list(zip(daylist,cumilative))
dfc=pd.DataFrame(data=dataset_cumilative, columns=['Day', 'Cumilative Miles Covered'])
dfc
```

### Let us plot this in chart

```python
dfc['Cumilative Miles Covered'].plot()

maxvalue = dfc['Cumilative Miles Covered'].max()

maxday = dfc['Day'][dfc['Cumilative Miles Covered'] == dfc['Cumilative Miles Covered'].max()].values

text = str(maxvalue) + " - " + maxday
plt.annotate(text, xy=(1, maxvalue), xytext=(8, 0), 
                 xycoords=('axes fraction', 'data'), textcoords='offset points')
#print("The most popular name")
dfc[dfc['Cumilative Miles Covered'] == dfc['Cumilative Miles Covered'].max()]

```
